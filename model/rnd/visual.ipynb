```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def visual(df: pd.DataFrame, sensor_positions: list[tuple[int, int]], points_df: pd.DataFrame):
    # Визуализация
    plt.figure(figsize=(8, 6))

    # Тепловая карта сигнала (RSSI)
    plt.scatter(df['x'], df['y'], color='red', s=100, label='Grid Points')

    plt.scatter(points_df['X_pred'], points_df['Y_pred'], color='green', s=100, label='Predicted Points')

    # Датчики
    for sensor in sensor_positions:
        plt.scatter(sensor[0], sensor[1], color='blue', edgecolor='black', s=200, marker='D', label='Sensor')

    # Настройки графика
    plt.title('Signal Visualization')
    plt.xlabel('X (m)')
    plt.ylabel('Y (m)')
    plt.grid(True, alpha=0.2)

    # Отображение
    plt.show()
```

```python
df = pd.DataFrame(actual_data, columns=actual_data.columns)
sens = list(sensors.itertuples(index=False, name=None))

visual(df=df,
                   sensor_positions=sens,
                   points_df=points_df)
```
